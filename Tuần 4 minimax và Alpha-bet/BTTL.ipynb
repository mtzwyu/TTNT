```python
# ==========================================
# BÀI 3: THUẬT TOÁN MINIMAX CHO TICTACTOE NxN
# ==========================================
import math

class MinimaxTicTacToe:
    def __init__(self, n=5):
        self.n = n
        self.board = [['_' for _ in range(n)] for _ in range(n)]
        self.player = 'X' # Máy (Max)
        self.opponent = 'O' # Người (Min)

    def print_board(self):
        for row in self.board:
            print(" | ".join(row))
            print("-" * (self.n * 4 - 1))
        print()

    def get_empty(self):
        return [(i, j) for i in range(self.n) for j in range(self.n) if self.board[i][j] == '_']

    # Hàm đánh giá trạng thái cơ bản (Dùng khi đạt giới hạn độ sâu)
    def evaluate_state(self):
        return 0 # Giả lập trạng thái cân bằng

    # Thuật toán Minimax lõi
    def minimax(self, depth, is_max):
        # Điều kiện dừng: Ở đây đơn giản hóa là hết độ sâu hoặc hết ô trống
        empty_cells = self.get_empty()
        if depth == 0 or len(empty_cells) == 0:
            return self.evaluate_state()

        if is_max:
            best_val = -math.inf
            for (i, j) in empty_cells:
                self.board[i][j] = self.player
                value = self.minimax(depth - 1, False)
                self.board[i][j] = '_'
                best_val = max(best_val, value)
            return best_val
        else:
            best_val = math.inf
            for (i, j) in empty_cells:
                self.board[i][j] = self.opponent
                value = self.minimax(depth - 1, True)
                self.board[i][j] = '_'
                best_val = min(best_val, value)
            return best_val

    def find_best_move(self, depth=3):
        best_val = -math.inf
        best_move = (-1, -1)
        for (i, j) in self.get_empty():
            self.board[i][j] = self.player
            move_val = self.minimax(depth - 1, False)
            self.board[i][j] = '_'

            if move_val > best_val:
                best_move = (i, j)
                best_val = move_val
        return best_move

# --- CHẠY THỬ BÀI 3 ---
print("--- TEST BÀI 3: MINIMAX 5x5 ---")
game_b3 = MinimaxTicTacToe(n=5)
# Đặt thử vài quân cờ
game_b3.board[2][2] = 'X'
game_b3.board[2][3] = 'O'
game_b3.print_board()

print("Đang suy nghĩ (Minimax)...")
# Khuyến cáo: Giữ depth nhỏ (2 hoặc 3) vì Minimax rất chậm trên bảng lớn
move_b3 = game_b3.find_best_move(depth=2)
print(f"Nước đi tốt nhất cho X là: {move_b3}")
```

```python
# ==========================================
# BÀI 4: THUẬT TOÁN ALPHA-BETA CHO TICTACTOE NxN
# ==========================================
import math

class AlphaBetaTicTacToe:
    def __init__(self, n=5):
        self.n = n
        self.board = [['_' for _ in range(n)] for _ in range(n)]
        self.player = 'X'
        self.opponent = 'O'

    def print_board(self):
        for row in self.board:
            print(" | ".join(row))
            print("-" * (self.n * 4 - 1))
        print()

    def get_empty(self):
        return [(i, j) for i in range(self.n) for j in range(self.n) if self.board[i][j] == '_']

    def evaluate_state(self):
        return 0

    # Thuật toán Alpha-Beta lõi
    def alphabeta(self, depth, alpha, beta, is_max):
        empty_cells = self.get_empty()
        if depth == 0 or len(empty_cells) == 0:
            return self.evaluate_state()

        if is_max:
            best_val = -math.inf
            for (i, j) in empty_cells:
                self.board[i][j] = self.player
                value = self.alphabeta(depth - 1, alpha, beta, False)
                self.board[i][j] = '_'

                best_val = max(best_val, value)
                alpha = max(alpha, best_val)
                if beta <= alpha:
                    break # Cắt tỉa nhánh Beta
            return best_val
        else:
            best_val = math.inf
            for (i, j) in empty_cells:
                self.board[i][j] = self.opponent
                value = self.alphabeta(depth - 1, alpha, beta, True)
                self.board[i][j] = '_'

                best_val = min(best_val, value)
                beta = min(beta, best_val)
                if beta <= alpha:
                    break # Cắt tỉa nhánh Alpha
            return best_val

    def find_best_move(self, depth=4):
        best_val = -math.inf
        best_move = (-1, -1)
        for (i, j) in self.get_empty():
            self.board[i][j] = self.player
            # Gọi hàm với alpha = -vô cực, beta = vô cực
            move_val = self.alphabeta(depth - 1, -math.inf, math.inf, False)
            self.board[i][j] = '_'

            if move_val > best_val:
                best_move = (i, j)
                best_val = move_val
        return best_move

# --- CHẠY THỬ BÀI 4 ---
print("--- TEST BÀI 4: ALPHA-BETA 5x5 ---")
game_b4 = AlphaBetaTicTacToe(n=5)
game_b4.board[2][2] = 'X'
game_b4.board[2][3] = 'O'
game_b4.print_board()

print("Đang suy nghĩ (Alpha-Beta)...")
# Có thể để depth lớn hơn Minimax một chút vì đã có cắt tỉa
move_b4 = game_b4.find_best_move(depth=3)
print(f"Nước đi tốt nhất cho X là: {move_b4}")
```

```python
# ==========================================
# BÀI 6: GIAO DIỆN ĐỒ HỌA CHO TICTACTOE
# ==========================================
import tkinter as tk
from tkinter import messagebox

class TicTacToeGUI:
    def __init__(self, root, n=3): # Có thể truyền n=5, n=10
        self.root = root
        self.root.title(f"Trò chơi Tic Tac Toe {n}x{n}")
        self.n = n
        self.current_player = "X"
        self.board = [["" for _ in range(n)] for _ in range(n)]
        self.buttons = [[None for _ in range(n)] for _ in range(n)]

        self.build_gui()

    def build_gui(self):
        # Khung chứa ma trận nút bấm
        frame = tk.Frame(self.root)
        frame.pack()

        for i in range(self.n):
            for j in range(self.n):
                btn = tk.Button(frame, text="", font=('Arial', 24, 'bold'),
                                width=4, height=2, bg='white',
                                command=lambda r=i, c=j: self.handle_click(r, c))
                btn.grid(row=i, column=j, padx=2, pady=2)
                self.buttons[i][j] = btn

        # Nút chơi lại
        reset_btn = tk.Button(self.root, text="Bắt đầu lại", font=('Arial', 14), command=self.reset_game)
        reset_btn.pack(pady=10)

    def handle_click(self, r, c):
        # Nếu ô trống thì mới cho đánh
        if self.board[r][c] == "":
            # Cập nhật dữ liệu và giao diện
            self.board[r][c] = self.current_player
            self.buttons[r][c].config(text=self.current_player,
                                      fg="blue" if self.current_player == "X" else "red")

            # Đổi lượt
            self.current_player = "O" if self.current_player == "X" else "X"

    def reset_game(self):
        self.current_player = "X"
        self.board = [["" for _ in range(self.n)] for _ in range(self.n)]
        for i in range(self.n):
            for j in range(self.n):
                self.buttons[i][j].config(text="", bg='white')

# --- CHẠY GIAO DIỆN ---
# Chú ý: Khi chạy cell này, một cửa sổ ứng dụng mới sẽ bật lên (Pop-up)
if __name__ == "__main__":
    # Đóng cửa sổ cũ nếu bạn chạy lại cell nhiều lần
    try:
        root.destroy()
    except:
        pass

    root = tk.Tk()
    # Bạn có thể thay n=3 thành n=5, n=10 tùy ý yêu cầu đề bài
    app = TicTacToeGUI(root, n=3)
    root.mainloop()
```
