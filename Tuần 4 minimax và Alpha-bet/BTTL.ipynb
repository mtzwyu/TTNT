```python
import copy
import math
import random
import numpy
import time # Added for AI thinking delay

X = "X"
O = 'O'
EMPTY = None
user = None
ai = None

def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """
    Returns player who has the next turn on a board.
    X always goes first.
    """
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)

    if x_count == o_count:
        return X # X's turn if counts are equal
    else:
        return O # O's turn if X has made one more move

def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res  = set()
    broad_len = len(board)
    for i in range(broad_len):
        for j in range(broad_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    i, j = action
    result_board[i][j] = curr_player
    return result_board

def get_horizontal_winner(board):
    # check horizontal
    for i in range(len(board)):
        if board[i][0] == board[i][1] == board[i][2] and board[i][0] is not EMPTY:
            return board[i][0]
    return None

def get_vertical_winner(board):
    #check vertically
    for j in range(len(board)):
        if board[0][j] == board[1][j] == board[2][j] and board[0][j] is not EMPTY:
            return board[0][j]
    return None

def get_diagonal_winner(board):
    #check diagonal
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] is not EMPTY:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] is not EMPTY:
        return board[0][2]
    return None

def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    horizontal_win = get_horizontal_winner(board)
    if horizontal_win:
        return horizontal_win

    vertical_win = get_vertical_winner(board)
    if vertical_win:
        return vertical_win

    diagonal_win = get_diagonal_winner(board)
    if diagonal_win:
        return diagonal_win

    return None

def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) is not None:
        return True
    for row in board:
        for cell in row:
           if cell == EMPTY:
                return False
    return True # All cells filled, and no winner, means a tie

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    else:
        return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
       return utility(state) # Fixed typo: utillity -> utility
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def miniMax(board):
    """
    Returns the optimal action for the current player on the board
    """
    current_player = player(board)
    optimal_move = None

        best_value = -math.inf # Renamed 'min' to 'best_value'
        for action in actions(board):
            value = minValue(result(board, action))
            if value > best_value:
                best_value = value
                optimal_move = action
    else: # O wants to minimize utility
        best_value = math.inf # Renamed 'max' to 'best_value'
        for action in actions(board):
            value = maxValue(result(board, action))
            if value < best_value:
                best_value = value
                optimal_move = action
    return optimal_move

if __name__ == "__main__":
    board = initial_state()
    print("Choose a player")
    user_symbol_input = input("Enter X or O: ").upper()

    if user_symbol_input == X:
        user = X
        ai = O
    else:
        user = O
        ai = X

    print(f"You are {user}, AI is {ai}.")
    print("\nInitial board:")
    for row in board:
        print([cell if cell is not None else '_' for cell in row])

    while True:
        game_over_status = terminal(board)
        if game_over_status:
            game_winner = winner(board)
            if game_winner is None:
                print("\nGame is a Tie!")
            else:
                print(f"\nPlayer {game_winner} wins!")
            break
        else:
            current_player_on_board = player(board) # Get the symbol of the current player (X or O)

            if current_player_on_board == user:
                print(f"\nYour turn ({user}). Enter the position to move (row, col):")
                while True:
                    try:
                        row_input = input("Row (0-2): ")
                        col_input = input("Col (0-2): ")
                        r = int(row_input)
                        c = int(col_input)

                        if (r, c) in actions(board):
                            board = result(board, (r, c))
                            break
                        else:
                            print("Invalid move: The chosen cell is not empty or coordinates are out of bounds. Try again.")
                    except ValueError:
                        print("Invalid input. Please enter numbers for row and column (0, 1, or 2).")
                    except (IndexError, TypeError):
                        print("Invalid coordinates. Row and column must be between 0 and 2. Try again.")
            else: # AI's turn
                print(f"\nAI's turn ({ai}). Thinking...")
                time.sleep(1) # Small delay for better UX
                move = miniMax(board)
                if move:
                    board = result(board, move)
                    print(f"AI chose to play at {move}")
                else:

                    print("AI couldn't find a move. This indicates an issue.")

            print("\nCurrent board:")
            for row in board:
                print([cell if cell is not None else '_' for cell in row])
```
