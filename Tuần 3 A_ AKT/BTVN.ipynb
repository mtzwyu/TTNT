```python
# BÀI TẬP VỀ NHÀ: BÀI TOÁN NGƯỜI GIAO HÀNG (TSP) SỬ DỤNG A*
import math
from heapq import heappush, heappop

class TSPGraph:
    def __init__(self, vertices, matrix):
        self.V = vertices       # Danh sách tên các đỉnh
        self.n = len(vertices)
        self.matrix = matrix    # Ma trận kề lưu trọng số

    def heuristic(self, unvisited):
        """
        Hàm Heuristic đơn giản: Lấy cạnh nhỏ nhất kết nối các đỉnh chưa thăm.
        (Để đảm bảo tính Admissible - h(n) <= h*(n))
        """
        if not unvisited: return 0
        min_edge = math.inf
        for u in unvisited:
            for v in unvisited:
                if u != v:
                    # Lấy index để tra ma trận
                    idx_u = self.V.index(u)
                    idx_v = self.V.index(v)
                    min_edge = min(min_edge, self.matrix[idx_u][idx_v])
        return 0 if min_edge == math.inf else min_edge

    def a_star_tsp(self, start):
        # State: (cost_f, cost_g, current_node, path_taken)
        pq = []
        start_idx = self.V.index(start)

        # Trạng thái ban đầu: path chỉ có [start]
        initial_path = [start]
        unvisited = [v for v in self.V if v != start]

        # h(n) ban đầu
        h = self.heuristic(unvisited)
        heappush(pq, (0 + h, 0, start, initial_path))

        print(f"Đang tìm đường đi TSP bắt đầu từ {start}...")

        while pq:
            f, g, current, path = heappop(pq)
            current_idx = self.V.index(current)

            # Nếu đã đi qua tất cả các đỉnh
            if len(path) == self.n:
                # Phải quay về đỉnh xuất phát
                return_cost = self.matrix[current_idx][start_idx]
                if return_cost > 0: # Giả sử có đường về
                    final_path = path + [start]
                    total_cost = g + return_cost
                    print(f"-> Chu trình tìm được: {' -> '.join(final_path)}")
                    print(f"-> Tổng chi phí: {total_cost}")
                    return total_cost, final_path

            # Khám phá các đỉnh chưa thăm
            unvisited = [v for v in self.V if v not in path]
            for nxt in unvisited:
                nxt_idx = self.V.index(nxt)
                weight = self.matrix[current_idx][nxt_idx]

                if weight > 0: # Có cạnh nối
                    new_g = g + weight
                    new_path = path + [nxt]
                    new_unvisited = [v for v in self.V if v not in new_path]

                    new_h = self.heuristic(new_unvisited)
                    new_f = new_g + new_h

                    heappush(pq, (new_f, new_g, nxt, new_path))

        print("Không tìm thấy chu trình đầy đủ!")
        return None

# --- CHẠY THỬ BÀI TOÁN NGƯỜI GIAO HÀNG ---
# Ma trận khoảng cách giữa các địa điểm (đồ thị đầy đủ)
# Chữ số 0 trên đường chéo chính, các số khác là khoảng cách
vertices = ['A', 'B', 'C', 'D']
distance_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp_solver = TSPGraph(vertices, distance_matrix)
print("--- GIẢI BÀI TOÁN TSP VỚI A* ---")
tsp_solver.a_star_tsp('A')
```
